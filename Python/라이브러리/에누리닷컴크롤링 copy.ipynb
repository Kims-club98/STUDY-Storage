```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
```

```python
# 드라이버 생성
driver = webdriver.Chrome()

# 접속
driver.get('https://www.enuri.com/search.jsp?keyword=%ED%95%98%EB%A6%BC%ED%8E%AB%ED%91%B8%EB%93%9C&from=list&cate=1642&tabType=1&sort=6&rental=N&delivery=N&pageGap=120&listGridType=1')
driver.maximize_window()
time.sleep(2)

click_path = [
'//*[@id="listBodyDiv"]/div[2]/div[4]/ul/li[1]/a/span[1]',
'//*[@id="chBF_0"]',
'//*[@id="chBF_1"]',
'//*[@id="chBF_2"]'
]

for xpath in click_path:
  try:
    # [수정 포인트 1] 매번 루프를 돌 때마다 현재 xpath에 해당하는 요소를 찾아서 button에 할당합니다.
    button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, xpath))
    )
    
    # [수정 포인트 2] 찾은 요소를 자바스크립트로 클릭합니다.
    driver.execute_script("arguments[0].click();", button)
    
    print(f"클릭 완료: {xpath}")
    
    # 필터가 적용되고 화면이 갱신되는 시간을 위해 약간의 대기 시간을 줍니다.
    time.sleep(2) 
  except Exception as e:
    print(f"클릭 중 에러 발생 ({xpath}): {e}")

# 3. 스크롤 시작
last_height = driver.execute_script("return document.documentElement.scrollHeight")
while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(2)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# 링크 수집
items = driver.find_elements(By.XPATH,'//div[@class="item__info"]//h3[@class="item__model"]/a')

detail_urls = []

for item in items:
  url = item.get_attribute('href')

  if url:
    if not url.startswith('http'):
      url = 'https://www.enuri.com'+url
    
    if url not in detail_urls:
      detail_urls.append(url)

print(f"✅ 정밀 수색 완료: 실제 사료 상품 {len(detail_urls)}개의 링크를 가져왔습니다.")

for i, link in enumerate(detail_urls):
    print(f"{i+1}번 상품 주소: {link}")

df = pd.DataFrame(detail_urls)
file_name = "detail_link.csv"
df.to_csv(file_name, index=False, encoding='utf-8-sig')

print(f'\n파일저장완료, 총 {len(detail_urls)}개 데이터 확보 완료. 파일명:{file_name}')


time.sleep(3)
driver.quit()
  
```

```python
import pandas as pd
import time
import os
import re # 파일명 세척을 위한 도구
import requests # 이미지 다운로드용
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import WebDriverException

# [1] 폴더 생성 (data/feed_img)
img_dir = 'data/feed_img'
if not os.path.exists(img_dir):
    os.makedirs(img_dir)

def restart_driver():
    print("🚀 브라우저를 시작합니다...")
    options = webdriver.ChromeOptions()
    return webdriver.Chrome(options=options)

def get_data(driver, xpath):
    try:
        # 수정 포인트: /text()를 제거하고 .text를 사용합니다.
        element = driver.find_element(By.XPATH, xpath)
        return element.text.strip()
    except:
        return "미기재"

# 초기 설정
detail_urls_df = pd.read_csv('./total_detail_link.csv', header=None)
url_list = detail_urls_df[0].tolist()
results = []
driver = restart_driver()

for i, url in enumerate(url_list):
    try:
        if i > 0 and i % 20 == 0:
            driver.quit()
            time.sleep(2)
            driver = restart_driver()

        print(f'[{i+1}/{len(url_list)}] 수집 중: {url}')
        driver.get(url)
        time.sleep(3)

        # 1. 상품명(item) 가져오기 - XPath 수정됨
        item_name = get_data(driver, '//*[@id="prod_summary_top"]/div/div[1]/div[1]')
        
        # 2. 파일명 세척 (파일명으로 쓸 수 없는 기호 _ 로 변경)
        clean_name = re.sub(r'[\/:*?"<>|]', '_', item_name)
        if clean_name == "미기재":
            clean_name = f"unknown_{i}"

        # 3. 이미지 추출 및 저장
        try:
            img_element = driver.find_element(By.XPATH, '//*[@id="prod_summary_left"]/div/div[1]/span/img')
            img_url = img_element.get_attribute('src')
            if img_url:
                img_res = requests.get(img_url)
                with open(f'{img_dir}/{clean_name}.jpg', 'wb') as f:
                    f.write(img_res.content)
                print(f"📸 이미지 저장 완료: {clean_name}.jpg")
        except:
            print(f"❌ 이미지 수집 실패: {clean_name}")

        # 4. 상세 정보 수집
        results.append({
            'item': item_name,
            'target': get_data(driver, '//*[@id="enuri_spec_table"]/dd[1]/table/tbody/tr[1]/td[1]'),
            'classification': get_data(driver, '//*[@id="enuri_spec_table"]/dd[1]/table/tbody/tr[1]/td[2]'),
            'age': get_data(driver, '//*[@id="enuri_spec_table"]/dd[1]/table/tbody/tr[2]/td[1]'),
            'form': get_data(driver, '//*[@id="enuri_spec_table"]/dd[1]/table/tbody/tr[2]/td[2]'),
            'processing_method': get_data(driver, '//*[@id="enuri_spec_table"]/dd[1]/table/tbody/tr[3]/td[1]'),
            'main_protein': get_data(driver, '//*[@id="enuri_spec_table"]/dd[1]/table/tbody/tr[3]/td[2]'),
            'allergy': get_data(driver, '//*[@id="enuri_spec_table"]/dd[1]/table/tbody/tr[4]/td[1]'),
            'char': get_data(driver, '//*[@id="enuri_spec_table"]/dd[1]/table/tbody/tr[4]/td[2]'),
            'certify': get_data(driver, '//*[@id="enuri_spec_table"]/dd[1]/table/tbody/tr[5]/td'),
            'crude_protein': get_data(driver, '//*[@id="enuri_spec_table"]/dd[2]/table/tbody/tr/td[1]'),
            'crude_fat': get_data(driver, '//*[@id="enuri_spec_table"]/dd[2]/table/tbody/tr/td[2]'),
            'url': url
        })

    except Exception as e:
        print(f'⚠️ 에러 발생: {e}')
        driver.quit()
        driver = restart_driver()
        continue

# 최종 저장
df = pd.DataFrame(results)
df.to_csv("result_of_ingredient.csv", index=False, encoding='utf-8-sig')
driver.quit()
print('🎉 모든 텍스트 및 이미지 수집 완료!')
```

```python
import pandas as pd
import time
import google.generativeai as genai
from selenium import webdriver
from selenium.webdriver.common.by import By

# 1. API 및 드라이버 설정
genai.configure(api_key="AIzaSyBfeakH9x1Gr4i4aEPaonc780E5t-Veg4E")
model = genai.GenerativeModel('gemini-1.5-flash')
driver = webdriver.Chrome() # 브라우저 실행

def get_info_from_ai(item_name, page_text):
    # 상세페이지 텍스트가 너무 길면 잘라서 전달 (토큰 절약)
    truncated_text = page_text[:2000] 
    
    prompt = f"""
    상품명: {item_name}
    상세페이지 내용: {truncated_text}
    
    위 내용을 읽고 다음 정보만 숫자로 추출해줘.
    1. 총 무게 (g 단위 숫자만)
    2. 총 칼로리 (kcal 숫자만, 내용에 없으면 0)
    
    답변 형식: 무게,칼로리 (예: 1600,3500)
    다른 설명은 절대 하지 마.
    """
    try:
        response = model.generate_content(prompt)
        return response.text.strip()
    except:
        return "0,0"

# 2. 데이터 불러오기
df = pd.read_csv('result_of_ingredient.csv')

# 3. 분석 루프
for i, row in df.iterrows():
    try:
        url = row['url'] # CSV에 url 열이 있어야 합니다.
        print(f"[{i+1}/{len(df)}] '{row['item']}' 분석 중...")
        
        # [수정] 실제로 페이지를 열어서 텍스트를 가져와야 합니다.
        driver.get(url)
        time.sleep(3) # 로딩 대기
        page_content = driver.find_element(By.TAG_NAME, 'body').text
        
        # [수정] 이름표(item)와 내용(page_content)을 함께 전달!
        ai_raw_data = get_info_from_ai(row['item'], page_content)
        print(f"   ㄴ AI 응답: {ai_raw_data}") # 진행 상황 확인용
        
        # 쉼표 기준 분리 및 저장
        weight_val, kcal_val = ai_raw_data.split(',')
        weight_g = float(weight_val)
        total_kcal = float(kcal_val)
        
        df.at[i, 'extracted_weight'] = weight_g
        df.at[i, 'total_kcal'] = total_kcal
        
        # 4. 1g당 칼로리 계산
        if weight_g > 0 and total_kcal > 0:
            df.at[i, 'kcal_per_1g'] = round(total_kcal / weight_g, 2)
        else:
            df.at[i, 'kcal_per_1g'] = "미기재"
            
    except Exception as e:
        print(f"⚠️ {i+1}번 상품 에러: {e}")
        continue
    
    time.sleep(1) # API 부하 방지

# 최종 저장
df.to_csv("result_final_complete.csv", index=False, encoding='utf-8-sig')
driver.quit()
print("✅ 모든 작업이 성공적으로 끝났습니다!")
```
