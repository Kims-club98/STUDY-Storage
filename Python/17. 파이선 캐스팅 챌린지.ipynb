x정수 값 1을 가진 변수를 생성합니다. <br>
부동소수점으로 변환하여 x저장합니다.a <br>
x문자열로 변환 하여 저장합니다.b <br>
인쇄 a, b, 및 해당 유형 <br>

```python
# Create an integer
x = 1
# Convert to float
a = float(x)
# Convert to string
b = str(x)
# Print values and types
print(a) # 1.0
print(b) # '1'
print(type(a)) # float
print(type(b)) # str
```
