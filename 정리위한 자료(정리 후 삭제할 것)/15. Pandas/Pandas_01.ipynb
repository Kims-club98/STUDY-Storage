### Pandas의 주요 구성 요소: Series, DataFrame, Index

  Serise - 1개의 column과 value의 구성의 데이터셋(1차 배열)(Set: 집합) <br>
  DataFrame - column x row 2차원의 데이터 셋(Serise의 모임) <br>
  Index - Series와 DataFrame의 고유한 Key값 <br>

#### Pandas의 주요 함수
```txt
데이터 로드 및 저장: read_csv(), read_html(), to_csv()
DataFrame 데이터 확인: df.shape, df.columns, df.head(), df.tail()
각 Column마다 Null 개수를 확인: df.isnull().sum()
특정 Column과 Row를 삭제하기: df.drop(['column'], axis=1)
```
